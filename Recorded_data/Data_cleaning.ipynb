# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
# Retina mode
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# Soham

```python
# Define the directory containing your CSV files and list them
directory = '/Users/umangshikarvar/Desktop/ML assignment/Soham'  # Replace with your directory path
csv_files = {
    "Walking": "Soham_Walking_1.csv",
    "Walking Upstairs": "Soham_Walking_Upstairs_1.csv",
    "Walking Downstairs": "Soham_Walking_Downstairs_1.csv",
    "Sitting": "Soham_Sitting_1.csv",
    "Standing": "Soham_Standing_1.csv",
    "Laying": "Shardul_Laying_1.csv"
}

# Plotting the data for each activity
plt.figure(figsize=(15, 18))

for i, (activity, filename) in enumerate(csv_files.items(), start=1):
    # Load the CSV file
    file_path = os.path.join(directory, filename)
    data = pd.read_csv(file_path)

    # Downsample the data to 50Hz
    data_downsampled = data.iloc[2000:6000:8].reset_index(drop=True)  # Keep every 8th sample to achieve 50Hz from 400Hz

    # Plot gFx, gFy, gFz, and TgF
    plt.subplot(len(csv_files), 1, i)
    plt.plot(data_downsampled.index, data_downsampled['gFx'], label='gFx', color='red')
    plt.plot(data_downsampled.index, data_downsampled['gFy'], label='gFy', color='green')
    plt.plot(data_downsampled.index, data_downsampled['gFz'], label='gFz', color='blue')
    plt.plot(data_downsampled.index, data_downsampled['TgF'], label='TgF', color='black')
    plt.title(f'{activity} (Downsampled to 50Hz)')
    plt.xlabel('Sample Number')
    plt.ylabel('Acceleration (g)')
    plt.legend()


# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
