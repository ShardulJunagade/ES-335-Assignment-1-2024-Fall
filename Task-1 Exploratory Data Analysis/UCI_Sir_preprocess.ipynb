# UCI - Dataset - Preprocessing and Saving

## Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.model_selection import train_test_split
```

## Get activity labels and subject ids

```python
activity_labels = {
    1: 'WALKING',
    2: 'WALKING_UPSTAIRS',
    3: 'WALKING_DOWNSTAIRS',
    4: 'SITTING',
    5: 'STANDING',
    6: 'LAYING'
}
print("Activity labels: ", activity_labels)


subject_train = pd.read_csv('../UCI HAR Dataset/train/subject_train.txt', delim_whitespace=True,header=None)
subject_train.columns = ['subject']

subject_test = pd.read_csv('../UCI HAR Dataset/test/subject_test.txt', delim_whitespace=True,header=None)
subject_test.columns = ['subject']

print("Shape of subject_train: ", subject_train.shape)
print("Shape of subject_test: ", subject_test.shape)
```

```python
# get feature names from the file features.txt
features = list()
with open('../UCI HAR Dataset/features.txt') as f:
    features = [line.split()[1] for line in f.readlines()]

print('No of Features: {}'.format(len(features)))
```

## Obtain the train and test data

```python
X_train = pd.read_csv('../UCI HAR Dataset/train/X_train.txt', sep=r'\s+', header=None)
y_train = pd.read_csv('../UCI HAR Dataset/train/y_train.txt', sep=r'\s+', header=None)

X_test = pd.read_csv('../UCI HAR Dataset/test/X_test.txt', sep=r'\s+', header=None)
y_test = pd.read_csv('../UCI HAR Dataset/test/y_test.txt', sep=r'\s+', header=None)

# Set column names
X_train.columns = features
X_test.columns = features
y_train.columns = ['activity']
y_test.columns = ['activity']

print("Training data shape: ",X_train.shape)
print("Training labels shape: ",y_train.shape)
print("Testing data shape: ",X_test.shape)
print("Testing labels shape: ",y_test.shape)
```

## Save activity and class wise data

### Training Data

```python
# Concatenate the subject and activity columns with the features
data = pd.concat([subject_train, y_train, X_train], axis=1)

os.makedirs('../Combined_UCI/train/', exist_ok=True)

# Save the X_train data activity-wise and subject-wise
for activity_id, activity in activity_labels.items():
    os.makedirs(f'../Combined_UCI/train/{activity}', exist_ok=True)
    
    for i in range(1, 31):
        activity_data = data[(data['activity'] == activity_id) & (data['subject'] == i)]
        
        # Check if there's any data for the current subject and activity
        if not activity_data.empty:
            activity_data_to_save = activity_data.drop(columns=['activity', 'subject'])
            activity_data_to_save.to_csv(f'../Combined_UCI/train/{activity}/subject_{i}.csv', index=False)
```

### Testing Data

```python
# Concatenate the subject and activity columns with the features
test_data = pd.concat([subject_test, y_test, X_test], axis=1)

os.makedirs('../Combined_UCI/test/', exist_ok=True)

# Save the X_test data activity-wise and subject-wise
for activity_id, activity in activity_labels.items():
    os.makedirs(f'../Combined_UCI/test/{activity}', exist_ok=True)
    
    for i in range(1, 31):
        activity_data = test_data[(test_data['activity'] == activity_id) & (test_data['subject'] == i)]
        
        # Check if there's any data for the current subject and activity
        if not activity_data.empty:
            activity_data_to_save = activity_data.drop(columns=['activity', 'subject'])
            activity_data_to_save.to_csv(f'../Combined_UCI/test/{activity}/subject_{i}.csv', index=False)
```

## MakeDataset Script

```python
folders = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
classes = {"WALKING":1,"WALKING_UPSTAIRS":2,"WALKING_DOWNSTAIRS":3,"SITTING":4,"STANDING":5,"LAYING":6}
```

```python
X_train = []
y_train = []
X_test = []
y_test = []

# Process the training data
dataset_dir = "../Combined_UCI/train"
for folder in folders:
    folder_path = dataset_dir + "/" + folder
    files = os.listdir(folder_path)

    for file in files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path, sep=",", header=0)
        X_train.append(df.values)
        y_train.extend([classes[folder]] * len(df))

X_train = np.vstack(X_train)
y_train = np.array(y_train)

# Process the test data
dataset_dir = "../Combined_UCI/test"

for folder in folders:
    folder_path = dataset_dir + "/" + folder
    files = os.listdir(folder_path)

    for file in files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path, sep=",", header=0)
        X_test.append(df.values)
        y_test.extend([classes[folder]] * len(df))

# Convert lists to NumPy arrays
X_test = np.vstack(X_test)
y_test = np.array(y_test)
```

```python
X_train.shape
```

```python
X = np.concatenate((X_train,X_test))
y = np.concatenate((y_train,y_test))

# split the data into training and testing sets. Change the seed value to obtain different random splits.
seed = 4
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=seed,stratify=y)

print("Training data shape: ",X_train.shape)
print("Testing data shape: ",X_test.shape)
```

## Save the numpy datasets

```python
os.makedirs('../SavedDataset/UCI_HAR', exist_ok=True)

np.save("../SavedDataset/UCI_HAR/UCI_X_train.npy",X_train)
np.save("../SavedDataset/UCI_HAR/UCI_X_test.npy",X_test)
np.save("../SavedDataset/UCI_HAR/UCI_y_train.npy",y_train)
np.save("../SavedDataset/UCI_HAR/UCI_y_test.npy",y_test)

print("Data saved successfully!")
```
